### Method:
1. Create empty DataFrame with specified columns
2. Create a function which will take in a country's name, and will get the specified info AND populate a row in the DataFrame
3. Clean the filled in DataFrame if needed
4. Convert the DataFrame to CSV or JSON???

```python
# Create an empty DataFrame with specified columns and 10 rows for 10 countries
df = pd.DataFrame(columns=['Name', 'Native Name', 'Capital(s)', 'Population', 'Languages', 'Region'])
df
```

```python
base_url = "https://restcountries.com/"

# Fill in the DataFrame with the given country's information
def get_country_info(name):
    url = f"{base_url}v3.1/name/{name}"
    response = requests.get(url)
 
    if response.status_code == 200:
        country_data = response.json() # if connection is successful, the data will be in json format and can be parsed as a dict

        # since the native names for the countries are in nested dicts of different languages, the below logic will standardize it
        native_names = country_data[0]["name"]["nativeName"]
        first_native_key = list(native_names.keys())[0]
        native_name = native_names[first_native_key]["official"]

        # Same format fix for languages as well
        languages = country_data[0]['languages']
        first_language_key = list(languages.keys())[0]
        language = languages[first_language_key]


        df.loc[len(df)] = [country_data[0]['name']['common'], native_name, country_data[0]['capital'][0], country_data[0]['population'], language, country_data[0]['region']]

    else:
        print(f"Failed to retrieve data for {name}. Status code: {response.status_code}")


# 10 countries to retrieve information for:

get_country_info("America")
get_country_info("Canada")
get_country_info("Pakistan")
get_country_info("India")
get_country_info("Korea")
get_country_info("China")
get_country_info("Japan")
get_country_info("Switzerland")
get_country_info("Italy")
get_country_info("Norway")


df
```

```python
# During all the testing, many duplicate entries were added to the DataFrame. Let's remove the test ones at the beginning

df = df.drop(df.index[0:21])
df
```

```python
df = df.iloc[1:] # need to drop one more oops
df
```

```python
df = df.reset_index(drop=True) # reset the index after dropping rows
df
```

### Last step: save data into clean CSV and JSON files

```python
df.to_csv("countries_clean.csv", index=False) #save the dataframe as a csv file
print("DataFrame saved to countries_clean.csv")

df.to_json("countries_clean.json", orient="index", indent=4)
print("DataFrame saved to countries_clean.json")
```
